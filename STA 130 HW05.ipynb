# 8. Complete the following assignment.

## 1. Problem Introduction:

This experiment is a variation of Fishers famous tea experiment, but instead of Dr. Bristol, we now have 80 students from STA130, each trying to determine whether milk or tea was poured first. The observed data shows that 49 out of 80 students guessed correctly, and we want to assess whether this result could be due to chance or if theres evidence that students can actually distinguish the order.

## 2. Relationship Between This Experiment and Fishers Original Experiment:

In Fisher's original experiment, the null hypothesis was that Dr. Bristol was guessing randomly, and the alternative hypothesis was that she could reliably distinguish between the two pouring methods. In our case, the population is different, and the number of trials is larger. Instead of a personalized ability, we are assessing a group of students who may not have prior experience in distinguishing this tea-milk difference.

## 3. Statements of the Null Hypothesis and Alternative Hypothesis:

1. Null Hypothesis H0: : The students are guessing randomly, meaning there is a 50% chance of correctly identifying the order of pouring. Formally, the proportion of correct guesses is p = 0.5
2. Alternative Hypothesis Ha: The proportion of students who can correctly identify the pouring order is greater than 50% (p > 0.5)
3. Informal Interpretation of H0: The students are no better at identifying the order of pouring than random guessing would be.

## 4. Quantitative Analysis:

We will perform a one-sided hypothesis test since we are testing whether the proportion of correct answers is greater than 0.5 (i.e., students are better than random guessing).

## 5. Methodology, Code, and Explanation:

To test the hypothesis, we will:
1. Simulate what random guessing looks like by generating a binomial distribution under the null hypothesis.
2. Calculate the p-value by comparing the observed test statistic (49 correct guesses out of 80) with the simulated distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Observed data
observed_correct = 49
total_students = 80
observed_proportion = observed_correct / total_students

# Null hypothesis: p = 0.5 (students are guessing)
p_null = 0.5

# Simulate the null distribution using binomial trials
np.random.seed(42)
num_simulations = 10000
simulated_data = np.random.binomial(total_students, p_null, num_simulations)

# Compute the p-value (proportion of simulations where the result is >= observed_correct)
p_value = np.mean(simulated_data >= observed_correct)

# Plot the simulation distribution for visualization
plt.hist(simulated_data, bins=range(min(simulated_data), max(simulated_data) + 1), alpha=0.75, label="Simulated data under H0")
plt.axvline(observed_correct, color='red', linestyle='dashed', linewidth=2, label=f'Observed ({observed_correct})')
plt.xlabel('Number of correct guesses')
plt.ylabel('Frequency')
plt.title('Simulated distribution under H0 with observed statistic')
plt.legend()
plt.show()

# Output p-value
p_value
```

## 6. Findings and Discussion:

After running the code:
1. We calculate the p-value, which tells us the probability of observing 49 or more correct guesses out of 80, assuming the students are randomly guessing.
2. If the p-value is small (typically less than 0.05), we can reject the null hypothesis and conclude that the students are better than random guessing.

## 7. Conclusion Regarding the Null Hypothesis:

Based on the p-value, we will decide whether to reject or fail to reject the null hypothesis. If the p-value is low, we reject H0 and conclude that there is evidence suggesting the students are able to correctly identify the order of pouring better than chance.
